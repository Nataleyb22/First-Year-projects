## .loc indexing Review

In pandas, we can index and filter based on specific categories of data that we are looking for using the .loc attribute.

Write code to import pandas and read in the United States COVID-19 Number of Cases Dataset, "time_series_covid19_us.csv"

```python
import pandas as pd
corona = pd.read_csv("time_series_covid19_us.csv")
corona.head()
```

When accessing data, we treat a dataframe as a dictionary where its columns are the keys and the data are the values.

Write code using the .loc attribute that displays just the cases from New York.

```python
df_ny = corona.loc[corona["Province/State"]=="New York"]
df_ny
```

Write code using the .loc attribute that displays all the states that had more than 100 cases on 3/14/20.

```python
df_morethan100 =corona["Province/State"].loc[corona["3/14/20"]>100]
df_morethan100
```
